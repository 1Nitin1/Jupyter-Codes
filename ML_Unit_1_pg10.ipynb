```python
#Page 4
# WHAT IS MACHINE LEARNING
# Machine Learning:
# It is the field of study that gives computers the ability to learn without being explicitly programmed.

# spam filter is given spam and ham examples to learn
# The examples that the system uses to learn is called training set.
# Each training example is called training instance (or smaple).
# The part of ML system that learns and predicts is called model.
```

```python
#Page 5
# WHY USE MACHINE LEARNING
# ML programs are shorter, easier to maintain and more accurate than traditional approach to many problems.
```

```python
#Page 6
# ML also shines in areas where the problem is too complex to implement using traditional approach or no such approach is known.
```

```python
#Page 7

# ML can also be used for data mining: It is the concept in which we inspect an ML model to draw insights by what it had learned.

# To summarize, ML is great for:
# 1. Problems for which traditional approaches are too long code.
# 2. Problems in which traditional approach doesn't give any valuable output.
# 3. Fluctuating Environment (ML model can be retrained on new data)
# 4. Data mining
```

```python
#Page 9
# TYPES OF ML SYSTEMS

#We can classify ML systems based on the following types:
# How they are supervised during training (supervised, unsupervised, semi-supervised, self-supervised, reinforcement).
# Whether or not they can learn incrementally on the fly (online vs batch learning).
# Whether it just compared new data points with existing ones or detects patterns to form a predictive model (instance-based vs model-based).
```

```python
#Page 10
# TRAINING SUPERVISION
# ML systems can be classified by the amount and type of supervision they get during training.
# supervised, unsupervised, semi-supervised, self-supervised, reinforcement

## SUPERVISED LEARNING
# In this, the training set includes the desired solutions called labels.
# A typical supervised learning task is classification.
# e.g., spam filter classifies messages as spam or ham.

# Another typical task is to predict a target numeric value of an object given its set of features.
# This is called regression.
# Some regression models can be used for classification and vice versa, e.g., logistic regression is used for classification. 
```
