```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
```

```python
# plotting a 2d vector
def plot_vec2d(x,**options):
    plt.grid()
    return plt.plot(x[0],x[1],**options)
```

```python
a = np.array([2,3])
```

```python
plot_vec2d(a,marker='o')
plt.show()
```

```python
#showing position vector arrow
def plot_vec2d_arr(y,x=[0,0],**options):
    return plt.arrow(x[0],x[1],y[0],y[1],head_width=0.1,length_includes_head=True,**options)
```

```python
plot_vec2d_arr(a,color='r')
plot_vec2d_arr((1,4),color='b')
plt.axis([0,5,0,5])
plt.grid()
plt.show()
```

```python
#plotting in 3d 
def plot_vec3d_m(x,**options):
    ax=plt.subplot(111,projection='3d')
    ax.grid()
    return ax.scatter(x[0],x[1],x[2],**options)
```

```python
b=np.array([1,2,3])
```

```python
plot_vec3d_m(b,s=10)
```

```python
# interactive plotly
def plot_vec3d_p(x,**options):
    df=pd.DataFrame({
        'x':[x[0]],
        'y':[x[1]],
        'z':[x[2]]
    })
    fig=px.scatter_3d(
        df,
        x='x',
        y='y',
        z='z',
        **options
    )
    fig.show()
```

```python
plot_vec3d_p(b)
```

```python
# Now lets draw position vector in 3d
def pos_3d(x,y,**options):
    ax=plt.subplot(111,projection='3d')
    ax.grid()
    ax.set_xlim([0,5])
    ax.set_ylim([0,5])
    ax.set_zlim([0,5])
    return ax.quiver(x[0],x[1],x[2],y[0],y[1],y[2],color='r',arrow_length_ratio=0.1,**options)
```

```python
pos_3d((0,0,0),b)
```

```python
# Calculating agle between becotors with dot product formula: a.b = ||a|| ||b|| cos(theta)
def ang(u,v):
    ang_rad=np.arccos(np.dot(u,v)/(norm(u)*norm(v)))
    return ang_rad*180/np.pi

ang(u,v)
```

```python
# Projecting a point on any axis
# projection of v on u = (u.v/(||u|| ||u||))*(u/||u||)
def proj(v,u):
    ax=plt.subplot(1,1,1)
    plot_vec2d_arr(u,color='r')
    plot_vec2d_arr(v,color='g')
    u_norm = u/norm(u)
    pro=np.dot(v,u_norm)*u_norm
    plot_vec2d_arr(pro)
    ax.plot([v[0],pro[0]],[v[1],pro[1]],color='b',linestyle=':')

proj([2,2],[1,5])
plt.axis([0,6,0,6])
plt.gca().set_aspect('equal')
plt.grid()
plt.show()
```

- diagonal: all elements except main diagonal are zeroes.
- scalar: diagonal matrix with all elements equal
- zero: all elements equal.
- singleton: only 1 element.
- identity: scalaar matrix with scalar=1.

```python
Q=np.array([[1,2,1,4],[2,5,3,5]])
Q
```

```python
def translate(P,Q):
    plt.gca().add_artist(Polygon(P.T,hatch='.',alpha=0.5,fc='r',ec='black'))
    plt.gca().add_artist(Polygon(Q.T,hatch='*',alpha=0.2,fc='b',ec='black'))
    plt.gca().add_artist(Polygon((P+Q).T,hatch='//',alpha=0.7,fc='g',ec='black'))
    for a,b in zip(P.T,Q.T):
        plot_vec2d_arr(b,a,color='r',linestyle=':')
    plt.axis([0,10,0,10])
    plt.grid()
    plt.gca().set_aspect('equal')
    plt.show()
```

```python
translate(P,Q)
```

```python
Q=np.array([[1,1,1,1],[2,2,2,2]])
translate(P,Q)
```

```python
def plot_transform(Pb,ratio):
    Pa=Pb*ratio
    plt.gca().add_artist(Polygon(Pb.T,alpha=0.5, ec='black', fc='g'))
    plt.gca().add_artist(Polygon(Pa.T, ec='black'))
    for a in Pa.T:
        plot_vec2d_arr(a,color='b',linestyle='--')
    for a in Pb.T:
        plot_vec2d_arr(a,color='r',linestyle=':')
    plt.axis([0,12,0,12])
    plt.grid()
    plt.show()
```

```python
plot_transform(P,2.6)
```

```python
# Matrix multiplication - Projection on axis
U = np.array([[1,0]])
U@P
```

```python
def plot_projection(U,P):
    U=U/norm(U)
    U_P = U@P
    axis_end = 100*U
    plot_vec2d_arr(axis_end[0],color='black')
    plt.gca().add_artist(Polygon(P.T,alpha=0.4,ec='black'))
    for vector, proj_coordinate in zip(P.T, U_P.T):
        print(vector, proj_coordinate)
        proj_point = proj_coordinate * U
        plt.plot(proj_point[0][0], proj_point[0][1], "ro", zorder=10)
        plt.plot([vector[0], proj_point[0][0]], [vector[1], proj_point[0][1]],
                 "r--", zorder=10)
    plt.gca().set_aspect('equal')
    plt.axis([0,5,-1,4])
    plt.grid()
    plt.show()
```

```python
plot_projection(U,P)
```
