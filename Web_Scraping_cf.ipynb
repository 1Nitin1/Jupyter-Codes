```python
import requests as re
from bs4 import BeautifulSoup
import regex as p
from typing import List
r=re.get("https://codeforces.com/contest/2234/standings")
s=BeautifulSoup(r.text,"lxml")
print(s.prettify())
```

```python
list1=[]
l=s.table.find("tr")
print(l)
```

```python
k=l.find_all("th")
print(k)
```

```python
list1.clear()
list2=[]
list1=[col.get_text(strip=True) for col in k]

j=s.table.find_all("tr")
i=1
while i<len(j):
    z=j[i]
    cols=z.find_all("td")
    list2.append([p.sub("\d{2}:\d{2}","",col.get_text(strip=True)) for col in cols])
    i+=1
list2.pop(len(list2)-1)
print(list1)
print(list2)
```

```python
import pandas as pd
c=list(zip(*list2))
data={}
i=0
for a in list1:
    if i==1:
        data[a]=list(c[1])
    else:
        data[a]=[int(e) if len(e)>0 else None for e in list(c[i])]
    i+=1
df=pd.DataFrame(data)
print(df)
df.to_excel("cf_scraped.xlsx",index=False,engine="openpyxl")
print("done")
```
