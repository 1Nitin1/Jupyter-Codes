```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('netflix.csv')
df.dropna(axis=0,inplace=True)
plt.figure(figsize=(5,5))
counts=df['type'].value_counts()
plt.bar(counts.index,counts.values,color='orange',width=0.5,label='Data')
plt.title('Movie vs TV show count')
plt.legend()
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('netbar.png',dpi=150,bbox_inches='tight')
plt.show()

plt.figure(figsize=(5,5))
cts=df['rating'].value_counts()
plt.pie(cts.values,autopct='%1.1f%%',labels=cts.index)
plt.legend()
plt.title('Rating distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('netpie.png',dpi=150,bbox_inches='tight')
plt.show()

plt.figure(figsize=(5,5))
ct=df['release_year'].value_counts()
plt.plot(ct.index,ct.values,label='Line Plot',color='red',linestyle='--',marker='s')
plt.legend()
plt.title('Release year vs number of releases')
plt.xlabel('year')
plt.ylabel('count')
plt.tight_layout()
plt.savefig('netline.png',dpi=150,bbox_inches='tight')
plt.show()

plt.figure(figsize=(5,5))
ct=df['country'].value_counts().head(8)
plt.barh(ct.index,ct.values,label='Horizontal bar Plot',color='blue',height=0.5)
plt.legend()
plt.title('Countries vs releases')
plt.xlabel('Country')
plt.ylabel('Shows')
plt.tight_layout()
plt.savefig('netbarh.png',dpi=150,bbox_inches='tight')
plt.show()

plt.figure(figsize=(5,5))
df_new=df[df['type']=='Movie'].copy()
df_new['new_time']=df_new['duration'].str.replace(' min','').astype(int)
ct=df_new['new_time'].value_counts()
plt.hist(df_new['new_time'],bins=30,label='Line Plot',color='red',edgecolor='black')
plt.legend()
plt.title('Movie Duration Distribution')
plt.xlabel('Time in minutes')
plt.ylabel('count of movies')
plt.tight_layout()
plt.savefig('nethist.png',dpi=150,bbox_inches='tight')
plt.show()

dfn=df.groupby(['release_year','type']).size().unstack().fillna(0)
fig, a=plt.subplots(1,2,figsize=(10,5))
a[0].plot(dfn.index,dfn['Movie'],label='Movies',color='blue')
a[0].set_title('Movies per year')
a[0].set_xlabel('Years')
a[0].set_ylabel('Movies')

a[1].plot(dfn.index,dfn['TV Show'],label='TV Shows',color='yellow')
a[1].set_title('TV Shows per year')
a[1].set_xlabel('Years')
a[1].set_ylabel('TV Shows')
fig.suptitle('Comparison of movie and tv show releases')
fig.tight_layout()
plt.savefig('netcomp.png',dpi=150,bbox_inches='tight')
plt.show()
```
