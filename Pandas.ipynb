```python
###Filtering and accessing
import pandas as pd
f={
    'Name':['Nitin','Nidhi','Niraj','Nitu','Manya','Niya'],
    'Age':[20,15,46,42,20,1],
    'Salary':[150000,5000,200000,5000,100000,5000]
}
df=pd.DataFrame(f)
print(df['Name'])
print(df[['Name','Age']])
print(df[(df['Age']>15) & (df['Salary']>5000)])
```

```python
#ADDING NEW COL : df['newcolname']=val
# or using df.insert(loc, 'colname',data)
import pandas as pd
f={
    'Name':['Nitin','Nidhi','Niraj','Nitu','Manya','Niya'],
    'Age':[20,15,46,42,20,1],
    'Salary':[150000,5000,200000,5000,100000,5000]
}
df=pd.DataFrame(f)
df['New_Salary']=df['Salary']*1.05
print(df)
df.insert(0,'Emp_id',[10,20,30,40,50,60])
print(df)
```

```python
#DELETING
#df.drop(columns=['colname',...],inplace=True/False)
import pandas as pd
f={
    'Name':['Nitin','Nidhi','Niraj','Nitu','Manya','Niya'],
    'Age':[20,15,46,42,20,1],
    'Salary':[150000,5000,200000,5000,100000,5000]
}
df=pd.DataFrame(f)
df.drop(columns=['Salary'],inplace=True)
print(df)
```

```python
###HANDLING MISSING VALUES 
# number -> nan, object -> None
#to check -> isnull()
import pandas as pd
import numpy as np
f={
    'Name':['Nitin',None,'Niraj','Nitu','Manya','Niya'],
    'Age':[20,15,np.nan,42,20,1],
    'Salary':[150000,5000,200000,5000,100000,5000]
}
df=pd.DataFrame(f)
print(df.isnull())
print(df.isnull().sum())

#to delete missing -> df.dropna(axis=0/1,inplace=True)
df.dropna(axis=0,inplace=False)
print(df)

#to replace missing -> df.fillna(val,inplace=True)
df.fillna({'Name':'Anonymous'},inplace=True)
print(df)
```

```python
#interpolate missing values
#df.interpolate(method='linear',axis=0/1,inplace=)
import pandas as pd
f={
    'Name':['Nitin','Nidhi','Niraj','Nitu','Manya','Niya'],
    'Age':[20,15,None,42,20,1],
    'Salary':[150000,5000,200000,5000,100000,5000]
}
df=pd.DataFrame(f)
df['Age']=df['Age'].interpolate(method='cubic',axis=0,inplace=False)
print(df)
```
