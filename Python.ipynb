```python
# defaultdict
document=["abcd","Nitin","Nitin","Nitin","abcd"]
#normal counting technique
word_ct={}
for word in document:
    if word in word_ct:
        word_ct[word]+=1
    else:
        word_ct[word]=1
print(word_ct)

#"forgiveness is better than permission" way
word_ct={}
for word in document:
    try:
        word_ct[word]+=1
    except KeyError:
        word_ct[word]=1;
print(word_ct)

#or use get logically
word_ct={}
for word in document:
    word_ct[word]=word_ct.get(word,0)+1
print(word_ct)

##USING DEFAULTDICT FROM COLLECTIONS
from collections import defaultdict
def f():
    return 1
ct=defaultdict(f) #takes a function input for default value or int,long,etc
for word in document:
    ct[word]+=1
print(ct)

#It can also be useful with list or dict or even your own functions:
d1=defaultdict(list)
d1["Nitin"].extend([100,200,300])
print(d1)
d2=defaultdict(dict)
d2["Nitin"]["Job"]="SDE"
print(d2)
d3=defaultdict(lambda: [0,0,0])
d3["abc"][2]=3
print(d3)
```

```python
#Object Oriented Programming
class CounterClicker:
    def __init__(self,count=0):
        self.count=count
    def __repr__(self):
        return f"CounterClicker(count={self.count})"
    def click(self,num_time=1):
        self.count+=num_time
    def read(self):
        return self.count
    def reset(self):
        self.count=0

clicker = CounterClicker(2)
assert clicker.read()==2,"check initialization"
clicker.click()
clicker.click()
assert clicker.read()==4,"check increment"
clicker.reset()
assert clicker.read()==0,"check reset"
clicker.click()
print(clicker)
```

```python
#making no reset clicker using inheritance
class NoResetClicker(CounterClicker):
    def reset(self):
        pass
clk2=NoResetClicker(5)
clk2.reset()
assert clk2.read()==5,"reset does nothing"
print(clk2)
```

```python
#Iterables and Generators
#Generator are usefuls when you need to iterate something once.
#generator created using yield or using for comprehensions
def gen_range(n):
    i=0
    while i<n:
        yield i
        i+=1
def natural_numbers():
    i=0
    while True:
        yield i
        i+=1
        
for i in gen_range(10):
    print(f"i: {i}",end=" - ")

evens_below_11=(i for i in range(11) if i%2==0)
for i in evens_below_11:
    print(f"i:{i}")
#evens_below_11 not usable now
data=natural_numbers()
evens=(x for x in data if x%2==0)
even_sq=(x**2 for x in evens)
# and so on... none of these does anything until iterated
#many time when we be iterating we also want indexes. for this we use enumerate fn.
l=[1,33,"abc",5.5]
#not pythonic
for i in range(len(l)):
    print(f"index={i} and value={l[i]}")
#also not pythonic
i=0
for x in l:
    print(f"index={i} and value={x}")
    i+=1
#This is pythonic: use enumerate- it converts into (index,value)
for i,val in enumerate(l):
    print(f"index={i} and value={val}")
```

```python
# ZIP AND ARGUMENT UNPACKING
#we can zip two or more iterables together. if iterables of different length
# , zip stops at smallest length end.
l1=[1,2,3,4]
l2=['a','b','c']
l3=[pair for pair in zip(l1,l2)]
print(l3)
#also, we can unzip using aesterisk called argument unpacking.
a,b=zip(*l3)
print(a,b)

#can use this in functions like:
def sum(a,b): return a+b
try:
    print(sum([2,3]))
except TypeError:
    print("Now trying with unpacking")
    print(f"sum = {sum(*[2,3])}")
```

```python
# ARGS AND KWARGS
#we create higher order functions as follows:
def doubler(f):
    def g(x):
        return 2*f(x)
    return g

def f(x):
    return x+1

l=doubler(f)
assert l(3)==8,"checking working"

#this work but if we pass more than 1 argument, it wont work since g only takes 1 input.
#thus, we use *args and **kwargs for list/tuple and dict inputs packing unpacking magic.
def doubler1(f):
    def g(*args,**kwargs):
        return 2*f(*args,**kwargs)
    return g

def f1(a,b,c):
    return a+b+c
la=[2,3]
lb={"c":4}
m=doubler1(f1)
assert m(*la,**lb)==18,"something is not right"
```
