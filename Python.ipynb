```python
# Module 1 : re
# can compile a regex pattern using re.compile. re.I means ignorecase
# methods are like match(), search(), findall(), sub()
# match() to get prefix match
# search() to get first occurence
# findall() for all occurences
# sub() to substitute all occurences
import re as regex
my_regex = regex.compile("[0-9]+",regex.I)
print(my_regex.match("123abc")) # or can use re.match(pattern, string) 
```

```python
#FUNCTIONS
#python functions are first class i.e.,
#we can assign them to variabes and pass thm into functions just like any other arguments
def double(x):
    return 2*x
def one(f):
    return f(1)
print(one(double))
my_double=double
print(one(my_double))
```

```python
#Creating lambda functions
z = lambda x: x+4
print(z(2))
```

```python
# Default argument passing
def name(n="Anonymous"):
    print(n)
name()    
```

```python
## STRINGS
# can create strings using '', "", ''' ''', """ """
# "\t" length is 1
# r"\t" means raw string so, len is 2
fname='Nitin'
lname='Baranwal'
name1= fname+" "+lname
name2="{0} {1}".format(fname,lname)
name3=f"{fname} {lname}"
print(name1,name2,name3)
```

```python
## EXCEPTIONS
try:
    print(0/0)
except ZeroDivisionError:
    print('Cannot divide by zero')
```

```python
## LISTS
list=[2,1,3,4,5]
list.sort()
print(list)
# methods like sum, len, etc
# can get elements by +ve or -ve indexing
```

```python
# slicing lists : list[start:stop(ex):jump]
k=list[1:5:2]
print(k)
rev=list[::-1]
print(rev)
```

```python
# checking membership using in
print(1 in [1,2,3])
# adding elements to a list: use extend or + or append
x=[1,2,3]
y=[4,5,6]
y.extend(x)
print(y+x)
y.append("a")
print(y)

# unpacking: _ if no need for the element
x,y=[1,2]
print(x,y)
_,b=[1,2]
print(b)
```

```python
## TUPLES : immutable
a=() # it is a tuple
b=set() # it is an empty set
c=(1) # not a tuple
d=(1,) # is a tuple
```
