```python
#Page 18

# BATCH VS ONLINE LEARNING

# BATCH LEARNING
# In this, the Ml system is incapable to learn incrementally, i.e., we train the model on all the data at once and launch into production.
# If now we want it to learn new data, we need to retrain the model on the whole data including the new one again.
# This cycle can be automated easily but, will cost too much for storing and managing the loads of data at once.
# In this, there is a possibility that the data becomes old and model rot or data drift happens.
# data drft/model rot happens when the performance of model worsens over time due to older data.
```

```python
#Page 19

# ONLINE LEARNING
# In this, system has the ability to learn incrementally.
# After training and launching into production, it is fed small training data as single or grouped instances( called mini-batches).
# Each learning step is fast and cheap so, system can learn about new data on the fly.
# Also we can train ML model for large data sets which can't fit into the main memory of our system at once(called out-of-core learning).
```

```python
#Page 20
# Also, a parameter associated with online learning is the learning rate.
# If learning rate is high, model will adapt to new data quickly but forget the old data.
# If you set a low learning rate, model will have more inertia and will learn slowly.
# So, tweak the learning rate according to your needs.

# Outliers : Non-representative data points are called outliers.

# One challenge to online learning algorithm is that, if bad data is fed to the system, the model performance will drop and client will notice.
# So, we need to monitor the input data closely mainly using anomaly detection.
# Also, we need to closely monitor the performance of the system too.
```
