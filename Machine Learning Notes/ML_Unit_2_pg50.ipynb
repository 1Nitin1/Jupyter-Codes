## Pipelines
1. It is a sequence of data processing components.
2. The components run asynchronically.
3. One component processes data, spits the output in data store then, the next component starts, takes processed data from data store and runs similar.
4. This is quite robust if any component breaks the downstream still works on the previous output of broken component.
5. Proper monitoring is needed so that broken components can be fixed before any performance issue arise.

## Select a Performance Measure
- We generally select a performance measure to get the cost.
- RMSE and MAE are such examples.
- We can use gamma-k norm.
- MAE is gamma-1
- RMSE is gamma-2
- gamma-0 gives number of non zero features.
- gamma-inf gives max absolute value of a target.
- as degree increases, the sensitivity to outliers increases.
