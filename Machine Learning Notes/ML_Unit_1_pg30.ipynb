```python
#Page 21
## INSTANCE BASED VS MODEL BASED LEARNING
# We have two ways to generalize the predictions of a Machine learning algorithm i.e., instance based and learning based.
```

```python
# Page 21
# Instance based
# in this, we have the training data but we dont train any model.
# Instead, we compare the new data with the trainig data and predict based on the similarity.
# for example, KNN i.e., k nearest neighbours just compares the distances between vectors and predicts result.
```

```python
# Page 22
# Model based'
# In this, we actually train a model on the training data.
# then predicts the target of new instances using this trained model.
```

```python
#Page 22
# Examples:
# suppose you want to predict is money makes people happy, so you download happiness data and gdp per capita data.
# happiness data: www.oecdbetterlifeindex.org
# gdp per capita data: ourworldindata.org
```

```python
#Pagw 23-24
# Now join these two data sets on the country column and sort by gdp per capita
# When we plot this data on a graph, we can see that it somewhat formas a linear fashion.
# Eg: happiness = θ1 + θ2 * GDP per capita
# Linear model has 2 model parameters - slope and intercept
# How to determine their values ? Using some performance measure 
# either utility function(or fitness function) which tells how good the model is 
# or the cost function which tells how bad a model is 
# typically we use cost function and try to minimize the cost.
```

```python
#Page 25
# Let's build a linear and a KNN model to test both model and instance based learning.
# problem: To find the happiness index of a country given it's per capita income.
# download and load data
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
root="https://github.com/ageron/data/raw/main/"
df=pd.read_csv(root+"lifesat/lifesat.csv")
print(df)
```

```python
#Page 25
#visualizing the data
sns.regplot(data=df,x="GDP per capita (USD)",y="Life satisfaction",ci=None)
```

```python
#page 25
# Training the model
x=df[["GDP per capita (USD)"]].values
y=df[["Life satisfaction"]].values
my_model = LinearRegression()
my_model.fit(x,y)
#lets check if gdp per capita was 20000 how much will be the life satisfaction
print(my_model.predict([[20000]]))
```

```python
#Page 26
# now for KNN,
from sklearn.neighbors import KNeighborsRegressor
predictor=KNeighborsRegressor(n_neighbors=3)
# lets predict for 41000 gdp per capita
# we see that New Zealand, UK and france are the nearest to 41000 so, answer should be their average : (7.3+6.8+6.5)/3 = 6.866667 
predictor.fit(x,y)
print(predictor.predict([[41000]]))
```
