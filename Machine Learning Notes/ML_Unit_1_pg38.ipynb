# Evaluating and fine tuning

## testing and Validation

We test model my splitting data into training and test data and then testing using the test data.
The error we get on the predictions of test data is called generalisation error.
If training error is low but generalisation error is high that means our model is overfitting.
Generally, we take 20 percent of data as test set but it also depends on the size of data.

## Hyperparameter Tuning and Model selection


1. For tuning and using the best hyperparameters, we need to repeatedly train and test the model on the training data and test data.
2. But, due to repeated testing on the same data, the model will adapt to that and as a result not perform good on real world data.
3. So, we split the training data further into another validation set and test on that.
4. Then, after choosing the best model retrain on train+validation set, we test it on test set and tune hyperparameters again if needed.
5. validation set is also called development or dev set.
6. This process is called hold out validation.
7. But we need to choose the optimal validation set size.
8. In order to avoid this, we use cross validation where we test the models of many small validation sets and average the accuracy.

## Data Mismatch
1. Lets say you are training a image recognition model that tell the species of flower image taken through mobile.
2. Now, for training this models you take training images from web.
3. You then take the images taken by mobile and split them into dev set and test set.
4. Now if the model performs bad on dev set, you won't know if it is because of image mismatch between web and mobile or due to tuning issue.
5. To get through this, we also make a train dev set by splitting the training set.
6. If model performs good on train-dev set then the issue is likely data mismatch.
7. Then, we can try to solve this problem by preprocessing mobile pics so that they look like web images.
